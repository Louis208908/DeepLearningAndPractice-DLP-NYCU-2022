```python
import numpy as np
import util
import NN

import matplotlib.pyplot as plt
```

```python
EPOCHS = 500

optis = [NN.optimizer('SGD', 0.01), NN.optimizer('momentum', 0.01, {'beta' : 0.9}), 
    NN.optimizer('adagrad', 0.01, {'epsilon' : 1e-8}), NN.optimizer('adam', 0.01, {'epsilon':1e-8, 'beta1':0.9, 'beta2':0.999})]
losses = []
for opti in optis:
    x, y = util.generate_linear(100)
    layers = [
        NN.FC_layer(2, 32, True),
        NN.relu(),
        NN.FC_layer(32, 32, True),
        NN.relu(),
        NN.FC_layer(32, 1, True),
        NN.sigmoid()
    ]
    DNN = NN.myNN(layers)
    loss = NN.MSE()

    losses.append([])

    for epoch in range(1, EPOCHS+1):
        y_pred = DNN.forward(x)
        loss_value = loss.forward(y, y_pred)
        losses[-1].append(loss_value)

        dy = loss.backward(y, y_pred)
        DNN.backward(dy, opti)


```

```python
plt.plot(losses[0], label='SGD')
plt.plot(losses[1], label='momentum')
plt.plot(losses[2], label='adagrad')
plt.plot(losses[3], label='adam')
plt.xlabel('Epochs')
plt.ylabel('Losses')
plt.title('Different Optimizers Comparison')
plt.legend()
plt.show()
```
